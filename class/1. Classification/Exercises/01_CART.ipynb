Use the wine dataset to train a CART decision tree and visualize the result. Can you do the same with the diabetes dataset? What happens? What accuracy do you obtain with each dataset?


## Wine dataset

Importing the libraries

```python
from sklearn.datasets import load_wine
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
```

Loading the data

```python
wine = load_wine()
df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
df['target'] = wine.target.astype(str)
df['target'] = "class_"  +  df['target']

#Get the data
X = df.drop('target', axis=1)
y = df['target']

df.head()
```

We need to split into train and test 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

Creation and training of the model

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

Now we make the predictions and compute the accuracy

```python
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print(f"Obtained accuracy: {acc}")
```

Finally, we visualize the tree

```python
tree.plot_tree(clf)
```

What hyperparameters do we have for our algorithm? You can consult them in the documentation of DecisionTreeClassifier. Change them and see what are the results. Why do you think that they affect the obtained results? Which configuration offers the best results?

The hyperparameters that we can tune are:

Description of the hyperparameters can be found [here](https://scikit-learn.org/dev/modules/generated/sklearn.tree.DecisionTreeClassifier.html).

Now we are going to change the hyperparameters and see the results of wine dataset again.

Loading the data

```python
wine = load_wine()
df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
df['target'] = wine.target.astype(str)
df['target'] = "class_"  +  df['target']

#Get the data
X = df.drop('target', axis=1)
y = df['target']

df.head()
```

We split the dataset into train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

As there are many hyperparameters, we are going to try the hyperparameters "automatically" using [GridSearchCV](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html).

```python
from sklearn.model_selection import GridSearchCV
```

Parameters we are going to try:

```python
param_grid = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 4, 6],
    'min_samples_leaf': [1, 2, 3],
    'max_features': [None, 'sqrt', 'log2']
}
```

Creation and training of the model

```python
clf = tree.DecisionTreeClassifier()
grid_search = GridSearchCV(clf, param_grid, cv=5,scoring='accuracy')
grid_search.fit(X_train, y_train)
```

See the best parameters and the obtained accuracy

```python
print(f"Obtained accuracy: {grid_search.best_score_}")
print(f"Best parameters: {grid_search.best_params_}")
```
